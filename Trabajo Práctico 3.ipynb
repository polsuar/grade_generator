3. Para 100 alumnos genere aleatoriamente las notas de 2 parciales.

a. Calcule el promedio que obtuvo cada estudiante.

b. Investigue y grafique un histograma con los promedios obtenidos

```
import matplotlib.pyplot as plt
import numpy as np


notas = np.random.randint(0,11,(100,2)) #con valores de 0 a 10, creo 100 filas con 2 columnas
prom = []

for i in range(100):
  prom.append((notas[i,0] + notas[i,1])/2)
  # print("El alumno nº",i+1,"con notas",notas[i,0],"y", notas[i,1],",tiene promedio",prom[i] )
print(prom)

plt.title('Promedio de alumnos')
plt.hist(prom,bins=50,alpha=1,edgecolor='black',linewidth=1)
plt.xlabel('Promedio') 
plt.ylabel('Numero de alumnos') 
plt.grid(True)
plt.gca().set_xticks(prom)
plt.show()
plt.clf()

```
